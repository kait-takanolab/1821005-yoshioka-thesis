```python
import cv2
import numpy as np
img = cv2.imread("kinkakuji.jpg")
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift = cv2.xfeatures2d.SIFT_create()
kp = sift.detect(gray,None)
img=cv2.drawKeypoints(gray,kp, None, flags=4)
cv2.imwrite('sift_kinkakuji.jpg',img)
```

```python
kp,des = sift.compute(gray,kp)
```
